```python
def retrieve(query: str, n_results: int = 3) -> dict:
    """
    Embeds the query and retrieves top-n relevant documents from ChromaDB.
    Returns a dict with:
      - context : formatted string with [Topic] labels (fed to LLM)
      - sources : list of topic names (for transparency)
      - raw     : raw ChromaDB results (for debugging)
    """
    # Encode the query into a vector
    query_embedding = embedder.encode(query).tolist()
 
    # Query ChromaDB for top n_results
    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=n_results,
        include=["documents", "metadatas", "distances"]
    )
 
    # Format results into a readable context string
    context_parts = []
    source_names  = []
 
    for i in range(len(results["documents"][0])):
        topic    = results["metadatas"][0][i]["topic"]
        text     = results["documents"][0][i]
        distance = results["distances"][0][i]
 
        context_parts.append(f"[{topic}]\n{text}")
        source_names.append(topic)
 
    context_string = "\n\n---\n\n".join(context_parts)
 
    return {
        "context": context_string,
        "sources": source_names,
        "raw":     results
    }
 
print("✅ Retrieval function defined!")
print("   Usage: retrieve('your legal question here', n_results=3)")
```

```python
import os
from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage, SystemMessage
 
# ── LLM Initialization ───────────────────────────────────────
llm = ChatOllama(
    model="phi3",
    temperature=0
)

# ── Agent Constants ───────────────────────────────────────────
FAITHFULNESS_THRESHOLD = 0.7   # Minimum acceptable faithfulness score
MAX_EVAL_RETRIES       = 2     # Maximum retries before accepting the answer
SLIDING_WINDOW         = 6     # Number of recent messages to keep in memory
TOP_K_RESULTS          = 3     # Number of ChromaDB chunks to retrieve
 
print("✅ LLM and constants initialized!")
print(f"   Model              : llama-3.3-70b-versatile")
print(f"   Faithfulness threshold : {FAITHFULNESS_THRESHOLD}")
print(f"   Max eval retries   : {MAX_EVAL_RETRIES}")
print(f"   Sliding window     : {SLIDING_WINDOW} messages")
print(f"   Top-K retrieval    : {TOP_K_RESULTS} chunks")
```

```python
def memory_node(state: CapstoneState) -> dict:
    """
    1. Appends the current question to messages history.
    2. Applies sliding window to prevent token overflow.
    3. Extracts user name if they introduce themselves.
    """
    question = state["question"]
    messages = state.get("messages", [])
    user_name = state.get("user_name", "")
 
    # Append current question to history
    messages.append({"role": "user", "content": question})
 
    # Apply sliding window — keep only the last N messages
    if len(messages) > SLIDING_WINDOW:
        messages = messages[-SLIDING_WINDOW:]
 
    # Extract user name if mentioned
    question_lower = question.lower()
    if "my name is" in question_lower:
        # Extract name after "my name is"
        try:
            name_part = question_lower.split("my name is")[1].strip()
            # Take first word as name
            extracted = name_part.split()[0].strip(".,!?")
            user_name = extracted.capitalize()
        except:
            pass  # If extraction fails, keep existing name
 
    return {
        "messages":  messages,
        "user_name": user_name,
    }
 
# ── Isolated Test for memory_node ────────────────────────────
print("Testing memory_node in isolation...")
 
mock_state_memory = {
    "question":    "My name is Rahul. What is Section 302?",
    "messages":    [],
    "user_name":   "",
    "route":       "",
    "retrieved":   "",
    "sources":     [],
    "tool_result": "",
    "answer":      "",
    "faithfulness": 0.0,
    "eval_retries": 0,
    "legal_domain": "",
}
 
result = memory_node(mock_state_memory)
print(f"\n✅ memory_node test passed!")
print(f"   Messages count : {len(result['messages'])}")
print(f"   Messages       : {result['messages']}")
print(f"   User name extracted : '{result['user_name']}'")
assert result["user_name"] == "Rahul", "Name extraction failed!"
assert len(result["messages"]) == 1, "Message not added!"
print("   All assertions passed ✅")
```

```python
def router_node(state: CapstoneState) -> dict:
    """
    Uses LLM to classify the question into one of three routes:
    - retrieve     : Question needs knowledge from the legal KB
    - tool         : Question needs current date/deadline calculation
    - memory_only  : Simple greeting, thanks, or follow-up needing no KB
    """
    question = state["question"]
    user_name = state.get("user_name", "")
 
    router_prompt = f"""You are a legal assistant router. Your job is to classify 
the user's question into exactly ONE of these three routes:
 
1. retrieve     - Use this when the question asks about any legal topic, law, 
                  IPC section, court procedure, rights, contracts, property, 
                  consumer rights, cyber law, bail, FIR, divorce, maintenance,
                  labour law, or any Indian legal matter.
 
2. tool         - Use this ONLY when the question asks about:
                  - Calculating a filing deadline or limitation period
                  - How many days remain before a legal deadline
                  - Current date-related legal calculations
 
3. memory_only  - Use this ONLY for:
                  - Simple greetings (hello, hi, thank you)
                  - Follow-up questions that reference the previous answer 
                    without needing new legal information
                  - Questions like "what did you just say?" or "repeat that"
 
User question: {question}
 
Rules:
- Reply with EXACTLY ONE word: retrieve, tool, or memory_only
- No explanation, no punctuation, just the single word
- When in doubt, choose retrieve"""
 
    response = llm.invoke([HumanMessage(content=router_prompt)])
    route = response.content.strip().lower()
 
    # Safety fallback — if LLM returns unexpected value, default to retrieve
    if route not in ["retrieve", "tool", "memory_only"]:
        route = "retrieve"
 
    return {"route": route}
 
# ── Isolated Test for router_node ────────────────────────────
print("Testing router_node with 4 different question types...")
 
router_tests = [
    ("What is the punishment for Section 302 IPC?",        "retrieve"),
    ("How many days do I have to file an appeal?",          "tool"),
    ("Hello, good morning!",                                "memory_only"),
    ("Is anticipatory bail available for murder cases?",    "retrieve"),
]
 
all_router_passed = True
for question, expected in router_tests:
    mock = {**mock_state_memory, "question": question}
    result = router_node(mock)
    status = "✅" if result["route"] == expected else "⚠️ "
    if result["route"] != expected:
        all_router_passed = False
    print(f"  {status} '{question[:45]}...' → {result['route']} (expected: {expected})")
 
print(f"\n{'✅ All router tests passed!' if all_router_passed else '⚠️  Some routes differ — review router prompt if needed'}")
print("Note: LLM routing may vary slightly — verify outputs make logical sense")
```

```python
def retrieval_node(state: CapstoneState) -> dict:
    """
    Embeds the user question and retrieves the top-K most
    relevant document chunks from ChromaDB.
    Returns formatted context string and source names.
    """
    question = state["question"]
 
    # Use the retrieve() helper defined in Day 2 (Cell 6)
    result = retrieve(question, n_results=TOP_K_RESULTS)
 
    # Detect legal domain from top source category
    top_source = result["sources"][0] if result["sources"] else ""
 
    return {
        "retrieved":    result["context"],
        "sources":      result["sources"],
        "legal_domain": top_source,
    }
 
# ── Isolated Test for retrieval_node ─────────────────────────
print("Testing retrieval_node in isolation...")
 
mock_retrieval = {**mock_state_memory, "question": "What are the rights of an arrested person?"}
result = retrieval_node(mock_retrieval)
 
print(f"\n✅ retrieval_node test passed!")
print(f"   Sources retrieved : {result['sources']}")
print(f"   Legal domain      : {result['legal_domain']}")
print(f"   Context length    : {len(result['retrieved'])} characters")
assert len(result["retrieved"]) > 100, "Retrieved context too short!"
assert len(result["sources"]) == TOP_K_RESULTS, f"Expected {TOP_K_RESULTS} sources!"
print("   All assertions passed ✅")
```

```python
def skip_retrieval_node(state: CapstoneState) -> dict:
    """
    Used when router decides 'memory_only'.
    Returns empty retrieved and sources so the answer_node
    does NOT leak context from the previous turn.
 
    ⚠️ CRITICAL: Must explicitly return '' and [] — not {}.
    Returning {} causes LangGraph to carry forward the previous
    turn's retrieved content into this turn's answer_node.
    This is exactly what MCQ Q12 in your exam tests!
    """
    return {
        "retrieved":    "",    # Explicit empty string — not None, not {}
        "sources":      [],    # Explicit empty list
        "legal_domain": "",
    }
 
# ── Isolated Test for skip_retrieval_node ────────────────────
print("Testing skip_retrieval_node in isolation...")
 
result = skip_retrieval_node(mock_state_memory)
print(f"\n✅ skip_retrieval_node test passed!")
print(f"   retrieved    : '{result['retrieved']}' (must be empty string)")
print(f"   sources      : {result['sources']} (must be empty list)")
assert result["retrieved"] == "", "retrieved must be empty string, not None or {}"
assert result["sources"]   == [], "sources must be empty list"
print("   All assertions passed ✅")
```

```python
from datetime import datetime, timedelta
import re
 
def tool_node(state: CapstoneState) -> dict:
    """
    Legal deadline calculator tool.
    Handles questions about:
    - Filing deadlines (how many days to file)
    - Limitation periods
    - Current date information
 
    ⚠️ Tools must NEVER raise exceptions — always return
    an error string instead. A crashing tool crashes the
    entire graph. This is what MCQ Q19 in your exam tests!
    """
    question = state["question"].lower()
 
    try:
        today = datetime.now()
        today_str = today.strftime("%d %B %Y")
 
        # ── Pattern 1: Days remaining from today ─────────────
        # e.g., "how many days left if I need to file within 90 days?"
        days_pattern = re.search(r'(\d+)\s*days?', question)
 
        if "how many days" in question or "days remaining" in question or "deadline" in question:
            if days_pattern:
                n_days = int(days_pattern.group(1))
                deadline_date = today + timedelta(days=n_days)
                result = (
                    f"Today's date is {today_str}. "
                    f"If your filing window is {n_days} days from today, "
                    f"your deadline falls on {deadline_date.strftime('%d %B %Y')}. "
                    f"You have {n_days} days remaining from today."
                )
            else:
                result = (
                    f"Today's date is {today_str}. "
                    f"Please specify the number of days in your question "
                    f"(e.g., 'How many days if I have 90 days to file?') "
                    f"so I can calculate your exact deadline."
                )
 
        # ── Pattern 2: Limitation period lookup ──────────────
        elif "limitation" in question or "time limit" in question:
            result = (
                f"Today's date is {today_str}. "
                f"Common limitation periods under the Limitation Act 1963:\n"
                f"• Recovery of money / breach of contract: 3 years\n"
                f"• Consumer complaint: 2 years from cause of action\n"
                f"• Recovery of immovable property: 12 years\n"
                f"• Criminal appeals: 30–90 days depending on court\n"
                f"• Civil appeal to High Court: 90 days from decree\n"
                f"Please consult a lawyer for your specific case timeline."
            )
 
        # ── Pattern 3: Current date query ────────────────────
        elif "today" in question or "current date" in question or "what date" in question:
            result = (
                f"Today's date is {today_str} "
                f"({today.strftime('%A')}). "
                f"The current year is {today.year}."
            )
 
        # ── Default: General date info ────────────────────────
        else:
            result = (
                f"Today's date is {today_str}. "
                f"For specific legal deadline calculations, please mention "
                f"the number of days or the type of case "
                f"(e.g., consumer complaint, civil suit, criminal appeal)."
            )
 
    except Exception as e:
        # ⚠️ NEVER raise — always return error string
        result = (
            f"I encountered an issue calculating the deadline: {str(e)}. "
            f"Please consult a legal professional for accurate deadline information."
        )
 
    return {"tool_result": result}
 
# ── Isolated Test for tool_node ──────────────────────────────
print("Testing tool_node with 3 different queries...")
 
tool_tests = [
    "What is today's date?",
    "How many days remaining if I need to file within 90 days?",
    "What is the limitation period for a consumer complaint?",
]
 
for q in tool_tests:
    mock = {**mock_state_memory, "question": q}
    result = tool_node(mock)
    print(f"\n  Query  : {q}")
    print(f"  Result : {result['tool_result'][:120]}...")
    assert len(result["tool_result"]) > 10, "Tool result too short!"
 
print(f"\n✅ tool_node test passed — no exceptions raised!")
```

```python
def answer_node(state: CapstoneState) -> dict:
    """
    Generates the final answer using:
    - Retrieved legal context (from retrieval_node)
    - Tool result (from tool_node, if any)
    - Conversation history (from memory_node)
 
    ⚠️ System prompt enforces strict grounding:
    The LLM must ONLY use information from the provided context.
    This is what keeps faithfulness scores high.
    """
    question     = state["question"]
    retrieved    = state.get("retrieved", "")
    tool_result  = state.get("tool_result", "")
    messages     = state.get("messages", [])
    user_name    = state.get("user_name", "")
    eval_retries = state.get("eval_retries", 0)
 
    # Build personalized greeting
    name_prefix = f"The user's name is {user_name}. Address them by name occasionally. " if user_name else ""
 
    # Escalation instruction for retries
    retry_instruction = ""
    if eval_retries > 0:
        retry_instruction = (
            f"\n\nIMPORTANT: This is retry attempt {eval_retries}. "
            f"Your previous answer scored below the faithfulness threshold. "
            f"Be more careful to use ONLY the information provided in the context. "
            f"Do not add any information not explicitly present in the context below."
        )
 
    # Build context section
    context_section = ""
    if retrieved:
        context_section = f"\n\nLEGAL KNOWLEDGE BASE CONTEXT:\n{retrieved}"
    if tool_result:
        context_section += f"\n\nTOOL RESULT (date/deadline calculation):\n{tool_result}"
 
    if not retrieved and not tool_result:
        context_section = "\n\n(No knowledge base context retrieved for this query.)"
 
    # Build conversation history string
    history_text = ""
    if len(messages) > 1:
        recent = messages[:-1][-4:]  # Last 4 messages excluding current
        history_text = "\n\nCONVERSATION HISTORY:\n"
        for msg in recent:
            role = "User" if msg["role"] == "user" else "Assistant"
            history_text += f"{role}: {msg['content'][:200]}\n"
 
    # System prompt with strict grounding rule
    system_prompt = f"""You are a knowledgeable and professional Indian legal assistant.
{name_prefix}
Your role is to help paralegals, junior lawyers, and citizens understand Indian law.
 
STRICT GROUNDING RULE:
- Answer ONLY using information provided in the LEGAL KNOWLEDGE BASE CONTEXT below.
- If the context contains the answer, provide it clearly and completely.
- If the question is outside the provided context or your knowledge base, say:
  "I don't have specific information about this in my knowledge base. 
   Please consult a qualified lawyer for this matter. 
   For legal aid, contact NALSA helpline: 15100."
- NEVER fabricate laws, sections, penalties, or legal procedures.
- NEVER use your general training knowledge to supplement the context.
- Always be professional, empathetic, and clear.
- For emergency legal situations, always mention: Police: 100, Legal Aid: 15100.
{retry_instruction}
{context_section}
{history_text}"""
 
    # Generate answer
    response = llm.invoke([
        SystemMessage(content=system_prompt),
        HumanMessage(content=question)
    ])
 
    return {"answer": response.content.strip()}
 
# ── Isolated Test for answer_node ────────────────────────────
print("Testing answer_node in isolation...")
 
# First retrieve some context
mock_answer = {
    **mock_state_memory,
    "question":   "What is the punishment for murder under IPC?",
    "user_name":  "Rahul",
    "eval_retries": 0,
}
# Get real retrieved context
retrieval_result = retrieval_node(mock_answer)
mock_answer["retrieved"] = retrieval_result["retrieved"]
mock_answer["sources"]   = retrieval_result["sources"]
 
result = answer_node(mock_answer)
print(f"\n✅ answer_node test passed!")
print(f"   Answer preview: {result['answer'][:300]}...")
assert len(result["answer"]) > 50, "Answer too short!"
print("   Assertion passed ✅")
```

```python
def eval_node(state: CapstoneState) -> dict:
    """
    Evaluates whether the generated answer is faithful to
    the retrieved context — i.e., it doesn't hallucinate.
 
    Faithfulness score:
    - 1.0 = answer uses ONLY information from the context
    - 0.7+ = acceptable (passes threshold)
    - <0.7 = fails threshold → answer_node retry triggered
    - If retrieved is empty (tool/memory route) → skip eval, score = 1.0
    """
    answer      = state.get("answer", "")
    retrieved   = state.get("retrieved", "")
    eval_retries = state.get("eval_retries", 0)
 
    # Skip faithfulness check if no KB context was retrieved
    # (tool or memory_only route — nothing to check against)
    if not retrieved.strip():
        return {
            "faithfulness":  1.0,
            "eval_retries":  eval_retries,
        }
 
    eval_prompt = f"""You are a strict legal content evaluator.
 
Your task: Score how faithful the ANSWER is to the CONTEXT provided.
 
Faithfulness means: Does the answer contain ONLY information that is 
explicitly present in the context? Or does it add facts, laws, section 
numbers, penalties, or procedures that are NOT in the context?
 
CONTEXT:
{retrieved[:2000]}
 
ANSWER TO EVALUATE:
{answer}
 
Scoring rules:
- 1.0 : Answer uses only information from the context, nothing added
- 0.8 : Answer mostly uses context, minor paraphrase differences
- 0.6 : Answer uses context but adds some outside information
- 0.4 : Answer partially uses context but adds significant outside info
- 0.2 : Answer barely uses context, mostly outside information
- 0.0 : Answer completely ignores the context
 
Reply with ONLY a decimal number between 0.0 and 1.0.
No explanation. No text. Just the number (e.g., 0.8)"""
 
    response = llm.invoke([HumanMessage(content=eval_prompt)])
 
    # Parse score safely
    try:
        score_text = response.content.strip()
        # Extract first number found
        numbers = re.findall(r'\d+\.?\d*', score_text)
        score = float(numbers[0]) if numbers else 0.5
        # Clamp to valid range
        score = max(0.0, min(1.0, score))
    except:
        score = 0.5  # Default to middle if parsing fails
 
    return {
        "faithfulness":  score,
        "eval_retries":  eval_retries + 1,
    }
 
# ── Isolated Test for eval_node ──────────────────────────────
print("Testing eval_node in isolation...")
 
mock_eval = {
    **mock_answer,
    "answer":      result["answer"],   # Use answer from answer_node test above
    "eval_retries": 0,
}
 
eval_result = eval_node(mock_eval)
print(f"\n✅ eval_node test passed!")
print(f"   Faithfulness score : {eval_result['faithfulness']}")
print(f"   Eval retries       : {eval_result['eval_retries']}")
print(f"   Threshold check    : {'PASS ✅' if eval_result['faithfulness'] >= FAITHFULNESS_THRESHOLD else 'FAIL — would retry'}")
assert 0.0 <= eval_result["faithfulness"] <= 1.0, "Score out of range!"
print("   Assertion passed ✅")
```

```python
def save_node(state: CapstoneState) -> dict:
    """
    Appends the assistant's final answer to the messages history.
    This is the last node before END.
    It ensures the answer is available in future turns via memory_node.
    """
    messages = state.get("messages", [])
    answer   = state.get("answer", "")
 
    # Append assistant response to history
    messages.append({"role": "assistant", "content": answer})
 
    return {"messages": messages}
 
# ── Isolated Test for save_node ──────────────────────────────
print("Testing save_node in isolation...")
 
mock_save = {
    **mock_eval,
    "messages": [{"role": "user", "content": "What is Section 302?"}],
    "answer":   "Section 302 IPC deals with punishment for murder...",
}
 
result = save_node(mock_save)
print(f"\n✅ save_node test passed!")
print(f"   Messages before save : 1")
print(f"   Messages after save  : {len(result['messages'])}")
print(f"   Last message role    : {result['messages'][-1]['role']}")
print(f"   Last message preview : {result['messages'][-1]['content'][:60]}...")
assert len(result["messages"]) == 2, "save_node did not append message!"
assert result["messages"][-1]["role"] == "assistant", "Role must be assistant!"
print("   All assertions passed ✅")
```

```python
def route_decision(state: CapstoneState) -> str:
    """
    Called after router_node by add_conditional_edges().
    Reads state.route and returns the name of the next node.
 
    ⚠️ Must be a STANDALONE function — LangGraph's
    add_conditional_edges() requires a callable.
    This is exactly what MCQ Q20 tests!
    """
    route = state.get("route", "retrieve")
 
    if route == "tool":
        return "tool"
    elif route == "memory_only":
        return "skip"
    else:
        return "retrieve"   # Default for "retrieve" + any unexpected value
 
 
def eval_decision(state: CapstoneState) -> str:
    """
    Called after eval_node by add_conditional_edges().
    Decides whether to retry answer_node or move to save_node.
 
    - faithfulness >= threshold            → PASS  → save
    - faithfulness <  threshold
      AND retries < MAX_EVAL_RETRIES       → RETRY → answer
    - faithfulness <  threshold
      AND retries >= MAX_EVAL_RETRIES      → ACCEPT → save
      (prevents infinite loops — MCQ Q3 tests this!)
    """
    faithfulness = state.get("faithfulness", 0.0)
    eval_retries = state.get("eval_retries", 0)
 
    if faithfulness >= FAITHFULNESS_THRESHOLD:
        return "save"
    elif eval_retries >= MAX_EVAL_RETRIES:
        return "save"       # Max retries hit → accept regardless of score
    else:
        return "answer"     # Score too low → retry
 
 
# ── Test both functions in isolation ─────────────────────────
print("Testing route_decision()...")
assert route_decision({"route": "retrieve"})    == "retrieve"
assert route_decision({"route": "tool"})        == "tool"
assert route_decision({"route": "memory_only"}) == "skip"
assert route_decision({"route": "garbage"})     == "retrieve"   # fallback
print("  ✅ All 4 cases correct\n")
 
print("Testing eval_decision()...")
assert eval_decision({"faithfulness": 0.85, "eval_retries": 1}) == "save"    # good score
assert eval_decision({"faithfulness": 0.50, "eval_retries": 0}) == "answer"  # retry
assert eval_decision({"faithfulness": 0.60, "eval_retries": 2}) == "save"    # max retries (MCQ Q3!)
print("  ✅ All 3 cases correct (including MCQ Q3 scenario)")
```

```python
from langgraph.graph import StateGraph, END
from langgraph.checkpoint.memory import MemorySaver
 
print("Assembling LangGraph — step by step...\n")
 
# Step 1: Instantiate graph with State class
graph = StateGraph(CapstoneState)
print("Step 1 ✅ StateGraph instantiated")
 
# Step 2: Add all 8 nodes
graph.add_node("memory",   memory_node)
graph.add_node("router",   router_node)
graph.add_node("retrieve", retrieval_node)
graph.add_node("skip",     skip_retrieval_node)
graph.add_node("tool",     tool_node)
graph.add_node("answer",   answer_node)
graph.add_node("eval",     eval_node)
graph.add_node("save",     save_node)
print("Step 2 ✅ All 8 nodes added")
 
# Step 3: Set entry point
graph.set_entry_point("memory")
print("Step 3 ✅ Entry point → memory")
 
# Step 4: Add fixed edges
graph.add_edge("memory",   "router")   # always
graph.add_edge("retrieve", "answer")   # always
graph.add_edge("skip",     "answer")   # always
graph.add_edge("tool",     "answer")   # always
graph.add_edge("answer",   "eval")     # always
graph.add_edge("save",     END)        # ⚠️ Most forgotten! MCQ Q6!
print("Step 4 ✅ Fixed edges added (including save → END)")
 
# Step 5: Add conditional edges
graph.add_conditional_edges(
    "router",
    route_decision,
    {
        "retrieve": "retrieve",
        "tool":     "tool",
        "skip":     "skip",
    }
)
 
graph.add_conditional_edges(
    "eval",
    eval_decision,
    {
        "answer": "answer",   # retry
        "save":   "save",     # accept
    }
)
print("Step 5 ✅ Conditional edges added (router + eval)")
 
# Step 6: Compile with MemorySaver
# MemorySaver persists graph state by thread_id — enables multi-turn memory
# MCQ Q14 tests the difference between MemorySaver and a plain Python list!
memory_saver = MemorySaver()
app = graph.compile(checkpointer=memory_saver)
print("Step 6 ✅ Compiled with MemorySaver\n")
 
print("=" * 55)
print("  ✅  GRAPH COMPILED SUCCESSFULLY!")
print("=" * 55)
print("  Nodes       : 8")
print("  Fixed edges : 6  (memory→router, retrieve/skip/tool→answer,")
print("                    answer→eval, save→END)")
print("  Conditional : 2  (router, eval)")
print("  Memory      : MemorySaver (thread_id based)")
```

```python
def ask(question: str, thread_id: str = "default") -> dict:
    """
    Sends a question to the compiled agent.
 
    Args:
        question  : User's legal question
        thread_id : Session ID — same ID = memory preserved across turns
                                 new ID  = fresh conversation
    Returns:
        Full result dict from the graph
    """
    config = {"configurable": {"thread_id": thread_id}}
 
    initial_state = {
        "question":     question,
        "messages":     [],
        "route":        "",
        "retrieved":    "",
        "sources":      [],
        "tool_result":  "",
        "answer":       "",
        "faithfulness": 0.0,
        "eval_retries": 0,
        "user_name":    "",
        "legal_domain": "",
    }
 
    result = app.invoke(initial_state, config=config)
 
    print(f"\n{'═'*62}")
    print(f"  Q: {question[:80]}")
    print(f"{'─'*62}")
    print(f"  Route        : {result.get('route',        'N/A')}")
    print(f"  Sources      : {result.get('sources',      [])}")
    print(f"  Faithfulness : {result.get('faithfulness', 'N/A')}")
    print(f"  Eval retries : {result.get('eval_retries', 0)}")
    if result.get("user_name"):
        print(f"  User name    : {result['user_name']}")
    print(f"{'─'*62}")
    print(f"  Answer:\n")
    for line in result.get("answer", "No answer generated").split("\n"):
        print(f"    {line}")
    print(f"{'═'*62}")
 
    return result
 
print("✅ ask() helper defined!")
print("   Usage: ask('Your legal question', thread_id='session1')")
```
